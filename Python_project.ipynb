WordsCloud - WordsCloud is a visual representation of the words that have the maximum frequency in the reviews section.

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
def removePunctuation(x):
    x = x.lower()
    x = re.sub(r'[^\x00-\x7f]',r' ',x)
    return re.sub("["+string.punctuation+"]", " ", x)
stops = set(stopwords.words("english"))
def removeStopwords(x):
    filtered_words = [word for word in x.split() if word not in stops]
    return " ".join(filtered_words)
def removeAmzString(x):
    return re.sub(r'[0-9]+ people found this helpful\. Was this review helpful to you Yes No', "", x)
reviews = [sent if type(sent)==str else "" for sent in df['title'].values]
reviews = [removeAmzString(sent) for sent in reviews]
reviews = [removePunctuation(sent) for sent in reviews]
mask = np.array(Image.open('C:/Users/ashis/Desktop/amazon.png'))
stopwords = set(STOPWORDS)
stopwords.add("shove")
stopwords.add("headphones")
wordcloud = WordCloud(background_color='#D5DBDB', stopwords=stopwords,mask=mask,max_words=100,
                      max_font_size=50,random_state=40).generate(str(reviews))
plt.figure(figsize=(15,20))
fig2 = plt.subplot2grid((4, 2), (2, 0), colspan=2,rowspan=2)
fig2.set_title("Customer Reviews cloud",fontsize=30,color='#4A235A')
fig2.imshow(wordcloud)
fig2.axis('off')
plt.show()
```

# Sentiment Analysis
The process of computationally identifying and categorizing opinions expressed in a piece of text, especially in order 
to determine whether the writers attitude towards a particular topic, product, etc., is positive, negative, or neutral. 
Sentiment Analysis with Python NLTK This is a demonstration of sentiment analysis using a NLTK . 
It can tell you whether it thinks the text being analysed expresses positive sentiment, negative sentiment, or if it's neutral.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
sentences=df1['title']
sid = SIA()
l1=[]
pol=0
for sentence in df1['title']:
    ss = sid.polarity_scores(str(sentence))
    l1.append(ss['compound'])
df4 = pd.DataFrame(l1)
df4['Products'] = df1['name'].copy()
df5 = df4.groupby(df4['Products'],as_index=False)[0].mean()
bestSeller = pd.DataFrame(df5.sort_values(0,ascending=False).head(5)['Products'])
bestSeller.index=[1,2,3,4,5]
bestSeller.index.name = 'RANK'
bestSeller
```
